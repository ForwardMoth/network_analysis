# Анализ постов

### Запрос к API. Метод wall.get

```python
def get_wall(group_id, offset=0):
    response = requests.get(BASIC_RESPONSE + 'wall.get', params={
        'access_token': TOKEN,
        'v': V,
        'domain': group_id, 
        'offset': offset, 
        'count': 100
    }).json()
    
    if 'response' in response:
        return response['response']    
```

###  Получаем посты с хэштегами

```python
def get_posts_with_hashtags(group_id): 
    text_posts_with_hashtags = []
    
    count = get_wall(group_id)['count'] 
    
    if count == 0:
        return members_list 
    # Вычитаем остаток от деления на 100, чтобы получить количество в сотнях
    max_offset = count - count % 100
    
    for offset in range(0, max_offset+1, 100):
        offset_posts = get_wall(group_id, offset)

        if offset_posts is not None:
            
            for item in offset_posts['items']:
                text = item['text']
                if '#' in text:
                    text_posts_with_hashtags.append(text)
            
    return text_posts_with_hashtags
```

```python
posts_iftes = get_posts_with_hashtags(iftes_group.group_name)
posts_icis = get_posts_with_hashtags(icis_group.group_name)
```

### Функция получения информации о хэштегах сообщества

```python
def get_hashtags(posts):
    hashtags_count = {}
    
    for text in posts:
        hashtags = [tag.strip("#") for tag in text.split() if tag.startswith("#")]
        
        for hashtag in hashtags:
            if hashtag in hashtags_count:
                hashtags_count[hashtag] += 1
            else:
                hashtags_count[hashtag] = 1
    
    return hashtags_count
```

```python
iftes_hashtags = get_hashtags(posts_iftes)
icis_hashtags = get_hashtags(posts_icis)
```

# Анализ количества постов за каждый час суток

```python
from datetime import datetime
```

```python
def get_post_activity(group_id):
    hours = {}
    
    count = get_wall(group_id)['count'] 
    max_offset = count - count % 100
    
    for offset in range(0, max_offset+1, 100):
        offset_posts = get_wall(group_id, offset)
        
        if offset_posts is not None:
            for item in offset_posts['items']:
                hour = datetime.fromtimestamp(item['date']).hour
            
                if hour in hours:
                    hours[hour] += 1 
                else:
                    hours[hour] = 1

    return hours
```

```python
iftes_post_activity = get_post_activity(iftes_group.group_name)
icis_post_activity = get_post_activity(icis_group.group_name)
```

# Построение социального графа

### Запрос к API. Метод friends.get

```python
def get_friends(user_id):
    response = requests.get(BASIC_RESPONSE + 'friends.get', params={
        'access_token': TOKEN,
        'v': V,
        'user_id': user_id
    }).json()
    
    if 'error' in response:
        return set()
    else: 
        return set(response['response']['items'])
```

### Получаем данные для графа

```python
def set_graph_data(first_users, second_users):
    graph = {}
    all_users = first_users | second_users 
    
    for user in all_users:
        friends = get_friends(user)
        common_friends = friends & all_users
        graph[user] = list(common_friends)
        
    return graph 
```

```python
graph_data = set_graph_data(set(iftes_group.members),set(icis_group.members))
```

### Построение графа

```python
import networkx as nx
```

```python
all_users = set(iftes_group.members) | set(icis_group.members)
```

```python
g = nx.Graph(directed=False)
for i in graph_data:
    g.add_node(i)
    for j in graph_data[i]:
        if i != j and i in all_users and j in all_users:
            g.add_edge(i, j)
            
nx.spring_layout(g)
nx.draw_networkx(g, with_labels=False, width=0.2, node_size=30)
```

```python
nodes=g.number_of_nodes()
print ("Кол-во узлов = ", nodes)
edges=g.number_of_edges()
print ("Кол-во ребер = ", edges)
```
